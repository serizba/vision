```python
import numpy as np
import skimage.io
import skimage
import matplotlib.pyplot as plt
import cv2
import numba
from tqdm import tqdm
```

```python
# Read model points
model_points = np.genfromtxt('model.txt').reshape([-1, 2])

num_p = model_points.shape[0]

H = np.zeros([5,3,3])

# Find homographies
for i in range(5):
    data_points = np.genfromtxt('data'+str(i+1)+'.txt').reshape([-1, 2])
    A_even = np.hstack([np.zeros((num_p,3)), -model_points, -np.ones((num_p,1)), data_points[:,1:]*model_points[:,:1], data_points[:,1:] * model_points[:,1:], data_points[:,1:]])
    A_odd = np.hstack([model_points, np.ones((num_p,1)), np.zeros((num_p,3)), -data_points[:,:1]*model_points[:,:1], -data_points[:,:1]*model_points[:,1:], -data_points[:,:1]])

    A = np.zeros((2*num_p, 9))
    A[np.arange(0, 2*num_p, 2)] = A_even
    A[np.arange(1, 2*num_p, 2)] = A_odd

    _, d, v = np.linalg.svd(A)
    H[i] = v[-1,:].reshape([3,3])
```

```python
# Normalization -> not done

# Decomposition of homography
A = np.zeros((2*5,6))
for i in range(5):
    v12 = [H[i,0,0]*H[i,1,0], H[i,0,0]*H[i,1,1]+H[i,0,1]*H[i,1,0], H[i,0,1]*H[i,1,1], H[i,0,2]*H[i,1,0]+H[i,0,0]*H[i,1,2], H[i,0,2]*H[i,1,1]+H[i,0,1]*H[i,1,2], H[i,0,2]*H[i,1,2]]
    v11 = [H[i,0,0]*H[i,0,0], H[i,0,0]*H[i,0,1]+H[i,0,1]*H[i,0,0], H[i,0,1]*H[i,0,1], H[i,0,2]*H[i,0,0]+H[i,0,0]*H[i,0,2], H[i,0,2]*H[i,0,1]+H[i,0,1]*H[i,0,2], H[i,0,2]*H[i,0,2]]
    v22 = [H[i,1,0]*H[i,1,0], H[i,1,0]*H[i,1,1]+H[i,1,1]*H[i,1,0], H[i,1,1]*H[i,1,1], H[i,1,2]*H[i,1,0]+H[i,1,0]*H[i,1,2], H[i,1,2]*H[i,1,1]+H[i,1,1]*H[i,1,2], H[i,1,2]*H[i,1,2]]
    A[2*i] = np.array(v12).T
    A[2*i+1] = (np.array(v11)-np.array(v22)).T

_, d, v = np.linalg.svd(A)
b = v[-1]

# Get internal parameters
y_0 = (b[1]*b[3] - b[0]*b[4]) / (b[0]*b[2]-b[1]**2)
l = b[5] - (b[3]**2 + y_0*(b[1]*b[3]-b[0]*b[4]))/b[0]
a_x = np.sqrt(l/b[0])
a_y = np.sqrt(l*b[0]/(b[0]*b[2]-b[1]**2))
s = -b[1]*a_x**2*a_y/l
x_0 = s*y_0/a_y - b[3]*a_x**2/l

print('x_0', x_0)
print('y_0', y_0)
print('a_x', a_x)
print('a_y', a_y)
print('s', s)
```

```python
def get_error(gt, pred):
    return np.mean(np.linalg.norm(gt-pred, axis=1))
```

```python
# Get external parameters of views and evaluate it
P_vector = np.zeros((5,3,4))

for view in range(5):

    K = np.array([[a_x, s, x_0],[0, a_y, y_0], [0,0,1]])
    K_inv = np.linalg.inv(K)

    n = 1/np.linalg.norm(K_inv @ H[view,:,0])

    r1 = n * K_inv @ H[view,:,0]
    r2 = n * K_inv @ H[view,:,1]
    r3 = np.cross(r1,r2)
    t = n * K_inv @ H[view,:,2]

    R = np.array([r1,r2,r3]).T

    M = np.hstack([R,t.reshape((-1, 1))])

    # Evaluate the calibration with view i
    P = K @ M
    P_vector[view] = P

    projected_p = P @ np.hstack([model_points, np.zeros((num_p,1)), np.ones((num_p,1))]).T

    projected_p = projected_p[:2] / projected_p[2:]
    
    # Get error of the calibration
    gt = np.genfromtxt('data'+str(view+1)+'.txt').reshape([-1, 2])
    print('AVERAGE ERROR', get_error(gt, projected_p.T))
    
    # Show the calibration
    plt.imshow(cv2.imread('CalibIm'+str(view+1)+'.pgm'))
    plt.scatter(projected_p[0], projected_p[1])
    plt.show()
```

```python
# Estimation of radial distortion (just from first view)
view = 0
P = P_vector[view]

p_tilde = P @ np.hstack([model_points, np.zeros((num_p,1)), np.ones((num_p,1))]).T
p_tilde = p_tilde[:2] / p_tilde[2:]
p_tilde = p_tilde.T
p = np.genfromtxt('data'+str(view+1)+'.txt').reshape([-1, 2])

r = np.sqrt((p_tilde[:,0] - x_0)**2 + (p_tilde[:,1] - y_0)**2)

A = np.array([r, r**2]).T
A = np.vstack([A,A])

b = np.hstack([(p[:,0]-p_tilde[:,0]) / (p_tilde[:,0]-x_0), (p[:,1]-p_tilde[:,1]) / (p_tilde[:,1]-y_0)])

k, _, _, _= np.linalg.lstsq(A,b,rcond=None)

print('Radial distortion coefficients: ', k)

# Correct for radial distortion
projected_p = (1 + k[0]*r + k[1]*r**2).reshape([-1,1]) * p_tilde

# Get error of the calibration
gt = np.genfromtxt('data'+str(view+1)+'.txt').reshape([-1, 2])
print('AVERAGE ERROR', get_error(gt, projected_p))


# Show the calibration
plt.imshow(cv2.imread('CalibIm'+str(view+1)+'.pgm'))
plt.scatter(projected_p[:,0], projected_p[:,1])
plt.show()
```
