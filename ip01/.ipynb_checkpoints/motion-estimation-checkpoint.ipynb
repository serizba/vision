```python
import numpy as np
import skimage.io
import skimage
import matplotlib.pyplot as plt
import cv2
import numba
from flow2color import *
from tqdm import tqdm
```

```python
# I(x,y)
I1, I2 = cv2.imread('yos08.pgm', 0), cv2.imread('yos09.pgm', 0)
# Apply Gaussian Blur
size = 7
I1, I2 = cv2.GaussianBlur(I1,(size, size),0), cv2.GaussianBlur(I2,(size, size),0)
# Stack the images -- I(x,y,t)
I = np.stack([I1, I2], axis=-1) / 255.0

gt = np.genfromtxt('yos_co.txt')

# Derivatives
Iy, Ix, It = np.gradient(I)
```

```python
@numba.jit(nopython=True, parallel=True)
def lucas_kanade(I, Ix, Iy, It, k):
    # Result matrix
    flow = np.zeros_like(I)
    
    ws = len(k) // 2
    
    for i in range(ws, I.shape[0]+ws):
        for j in range(ws, I.shape[1]+ws):
            Ix2  = (k * (Ix[i-ws:i+ws+1, j-ws:j+ws+1]**2)).sum()
            IxIy = ((k * Ix[i-ws:i+ws+1, j-ws:j+ws+1]*Iy[i-ws:i+ws+1, j-ws:j+ws+1])).sum()
            Iy2  = (k * (Iy[i-ws:i+ws+1, j-ws:j+ws+1]**2)).sum()
            IxIt = ((k * Ix[i-ws:i+ws+1, j-ws:j+ws+1]*It[i-ws:i+ws+1, j-ws:j+ws+1])).sum()
            IyIt = ((k * Iy[i-ws:i+ws+1, j-ws:j+ws+1]*It[i-ws:i+ws+1, j-ws:j+ws+1])).sum()
            if np.linalg.matrix_rank(np.array([[Ix2, IxIy], [IxIy, Iy2]])) == 2:
                flow[i-ws,j-ws] = np.linalg.solve(np.array([[Ix2, IxIy], [IxIy, Iy2]]), np.array([-IxIt, -IyIt]))
            else:
                flow[i-ws, j-ws] = 0
    return flow
```

```python
def flow2image(flow):
    res = np.zeros((flow.shape[0], flow.shape[1], 3), np.uint8)
    for i in range(flow.shape[0]):
        for j in range(flow.shape[1]):
            res[i,j] = np.uint8(flow2color(flow[i,j,0], flow[i,j,1]))
    return res
```

```python
def epe(gt, y):
    return np.sqrt((y[:, 0]-gt[:,0])**2 + (y[:, 1]-gt[:,1])**2).mean()
def aae(gt, y):
    return np.rad2deg(np.arccos((y[:, 0] * gt[:, 0] + y[:, 1] * gt[:, 1] + 1) / np.sqrt((y[:, 0]**2 + y[:, 1]**2 + 1) * (gt[:, 0]**2 + gt[:, 1]**2 + 1))).mean())
```

```python
# Try Lucas Kanade with different windows size and sigmas for gaussian kernel


i = 0

window_sizes = list(range(11, 41, 2))
sigmas       = [-1, 1, 2, 3]

err   = np.zeros(len(window_sizes) * len(sigmas))
flows = np.zeros((len(window_sizes) * len(sigmas), I.shape[0], I.shape[1], 2))

for ws, s in tqdm([(ws, s) for ws in window_sizes for s in sigmas]):
    k = cv2.getGaussianKernel(ksize=ws,sigma=s)
    k = k @ k.T
    flows[i] = lucas_kanade(I, *[np.pad(Ii[:,:,0], ((ws//2,ws//2), (ws//2,ws//2)), 'reflect') for Ii in [Ix, Iy, It]], k)
    err[i] = aae(gt, flows[i].reshape((-1, 2)))
    i += 1
```

```python
print(np.rad2deg(err.min()))
plt.imshow(flow2image(flows[err.argmin()]))
plt.show()
cv2.imwrite('lucas.png', cv2.cvtColor(flow2image(flows[err.argmin()]),cv2.COLOR_RGB2BGR))
```

```python
@numba.jit(nopython=True)
def sor(flow, Ix, Iy, It, alpha):
    # Result matrix
    # flow = np.zeros((Ix.shape[0], Ix.shape[1], 2))
    
    w = 1.99
    
    ialpha = 1.0 / alpha
        
        # oldflow = flow.copy()
    
    # Iterate over u
    for i in range(Ix.shape[0]):
        for j in range(Ix.shape[1]):
            neigh = (flow[i-1, j, 0] if i > 0 else 0) + (flow[i, j-1, 0] if j > 0 else 0.0) + (flow[i, j+1, 0] if j<Ix.shape[1] else 0.0) + (flow[i+1, j, 0] if i<Ix.shape[0] else 0.0)
            nu = (ialpha * It[i, j, 0] * Ix[i, j, 0] - neigh + ialpha * Ix[i, j, 0] * Iy[i, j, 0] * flow[i, j, 1]) / (-4 - ialpha * Ix[i, j, 0] ** 2) 
            flow[i, j, 0] = (1.0-w) * flow[i, j, 0] + w * nu

    # Iterate over v
    for i in range(Ix.shape[0]):
        for j in range(Ix.shape[1]):
            neigh = (flow[i-1, j, 1] if i > 0 else 0) + (flow[i, j-1, 1] if j > 0 else 0.0) + (flow[i, j+1, 1] if j<Ix.shape[1] else 0.0) + (flow[i+1, j, 1] if i<Ix.shape[0] else 0.0)
            nv = (ialpha * It[i, j, 0] * Iy[i, j, 0] - neigh + ialpha * Ix[i, j, 0] * Iy[i, j, 0] * flow[i, j, 0]) / (-4 - ialpha * Iy[i, j, 0] ** 2) 
            flow[i, j, 1] = (1.0-w) * flow[i, j, 1] + w * nv
                
        # print(it, aae(gt, flow.reshape((-1, 2))), ((oldflow-flow)**2).mean())
    return flow

def cuernos(Ix, Iy, It, alpha):
    flow = np.zeros((Ix.shape[0], Ix.shape[1], 2))
    err = 20
    #while abs(err -  aae(gt, flow.reshape((-1, 2)))) > 0.00001:
    for i in range(100000): 
        err = aae(gt, flow.reshape((-1, 2)))
        flow = sor(flow, Ix, Iy, It, alpha)
        if abs(err -  aae(gt, flow.reshape((-1, 2)))) < 0.0001:
            print(aae(gt, flow.reshape((-1, 2))))
            break
    return flow
```

```python
flow = cuernos(Ix, Iy, It, 1.0 / 125.0)
cv2.imwrite('cuernos.png', cv2.cvtColor(flow2image(flow),cv2.COLOR_RGB2BGR))
```
